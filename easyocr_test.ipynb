```python
import fitz  # PyMuPDF
from pdfminer.high_level import extract_text
import easyocr
import os

# EasyOCR 모델 초기화
reader = easyocr.Reader(['en', 'ko'])  # 다국어 가능

def extract_text_from_pdfminer(pdf_path):
    """PDF에서 텍스트만 추출 (pdfminer.six)"""
    return extract_text(pdf_path)

def extract_images_with_ocr_from_pdf(pdf_path, dpi=200, pages_to_ocr=None):
    """
    PDF 페이지를 이미지로 렌더링한 후 EasyOCR로 텍스트 추출
    :param pdf_path: PDF 파일 경로
    :param dpi: 렌더링 해상도
    :param pages_to_ocr: OCR 수행할 페이지 인덱스 리스트 (None이면 전체)
    :return: 딕셔너리 {페이지번호: OCR 텍스트}
    """
    doc = fitz.open(pdf_path)
    ocr_results = {}

    for page_num in range(len(doc)):
        if pages_to_ocr and page_num not in pages_to_ocr:
            continue

        page = doc.load_page(page_num)
        mat = fitz.Matrix(dpi / 72, dpi / 72)  # 해상도 변경
        pix = page.get_pixmap(matrix=mat)

        image_path = f"./page_{page_num + 1}.png"
        pix.save(image_path)

        # OCR 수행
        result = reader.readtext(image_path, detail=0)
        ocr_results[page_num + 1] = result

        # os.remove(image_path)  # 임시 파일 삭제

    return ocr_results

def process_pdf(pdf_path):
    """
    전체 파이프라인: 텍스트 + OCR
    """
    text_from_pdfminer = extract_text_from_pdfminer(pdf_path)
    ocr_from_images = extract_images_with_ocr_from_pdf(pdf_path)

    return {
        "text": text_from_pdfminer,
        "ocr": ocr_from_images
    }
```

```python
import fitz  # PyMuPDF
from pdfminer.high_level import extract_text
import easyocr
import os

reader = easyocr.Reader(['en', 'ko'])  # OCR 언어 설정

def extract_text_from_pdfminer(pdf_path):
    """텍스트만 추출 (pdfminer)"""
    return extract_text(pdf_path)

def get_pages_with_images(pdf_path):
    """이미지가 포함된 페이지 번호 리스트 반환 (0부터 시작)"""
    doc = fitz.open(pdf_path)
    image_pages = []
    for i, page in enumerate(doc):
        if page.get_images(full=True):  # 이미지가 하나라도 있으면
            image_pages.append(i)
    return image_pages

def extract_images_with_ocr_from_pdf(pdf_path, pages_to_ocr, dpi=200):
    """특정 페이지를 이미지로 렌더링 후 OCR 수행"""
    doc = fitz.open(pdf_path)
    ocr_results = {}

    for page_num in pages_to_ocr:
        page = doc.load_page(page_num)
        mat = fitz.Matrix(dpi / 72, dpi / 72)
        pix = page.get_pixmap(matrix=mat)

        image_path = f"page_{page_num + 1}.png"
        pix.save(image_path)

        result = reader.readtext(image_path, detail=0)
        ocr_results[page_num + 1] = result

        os.remove(image_path)

    return ocr_results

def process_pdf(pdf_path):
    text = extract_text_from_pdfminer(pdf_path)
    image_pages = get_pages_with_images(pdf_path)
    ocr = extract_images_with_ocr_from_pdf(pdf_path, image_pages)
    return {"text": text, "ocr": ocr}
```
